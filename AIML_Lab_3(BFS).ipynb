```python
from collections import deque

queue = deque()
queue.append(start)
visited = set([start])
parent = {start: None}
step = 0

print("Initial queue:", list(queue))
print("Initial visited:", visited)
print("Initial parent :", parent)

# -------------------------------
# 4) BFS Loop
# -------------------------------
# BFS Loop Condition
while queue:
    # Pause for Step-by-Step Execution

    # Tracks BFS iteration number.Increment Step Counter
    step += 1

    # Remove Node from Queue (Core BFS Action)
    node = queue.popleft()

    print("\nSTEP", step) # Print Current Step State
    print("Popped node:", node)
    print("Queue after pop:", list(queue))

    # Goal check
    if node == goal:
        print("Goal found! Stop BFS.")
        break

    # Explore neighbors, Retrieves neighbors of current node ;Iterates through each neighbor
    # Example:If node = 'A' → neighbors are B, C
    for neighbor in graph[node]:

        print(neighbor)

    # Check if Neighbor is Already Visited,Prevents revisiting nodes,Avoids infinite loops
    # Ensures each node is processed once
        if neighbor not in visited:

            # Mark Neighbor as Visited
            visited.add(neighbor)

            # store how we reached this neighbor
            parent[neighbor] = node

            # Adds neighbor to end of queue,Preserves FIFO order,Ensures level-by-level exploration
            queue.append(neighbor)

            # Print Updated State
            print("  Added to visited:", neighbor)
            print(f"  parent[{neighbor}] = {node}")
            print("  Added to queue:", neighbor)

    print("Queue now:", list(queue))
    print("Visited now:", visited)
# BFS Completion
print("\nBFS ended.")
print("Final visited:", visited)
print("Final parent :", parent)
```

```python
#    Defines a function named reconstruct_path
#    If BFS never reached goal, goal won’t exist in parent
def reconstruct_path(parent, start, goal):
    # If BFS never reached goal, goal won’t exist in parent, Return None meaning “no path found”
    if goal not in parent:
        return None

    path = [] #  path will store the final sequence of nodes

    current = goal# Start from the goal and walk backward

    # Move backward: goal -> parent[goal] -> ... -> start
    while current is not None:
        path.append(current)
        current = parent[current]

    # Right now path is reversed (goal to start), so reverse it
    path.reverse()
    return path
```

```python
path = reconstruct_path(parent, start, goal)

print("\nShortest path from", start, "to", goal, ":")
print(path)
```
