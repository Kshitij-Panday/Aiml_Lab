```python
# Create the graph as an adjacency list (dictionary)

graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': [],
    'F': []
}

# Define START and GOAL nodes
start = 'A'
goal  = 'F'
print(graph)
```

```python
print("Start node:", start)
print("Goal node:", goal)
```

```python
# Initialize DFS data structures
stack = [start]
visited = set([start])
parent = {start: None}
step = 0

# Print initial state
print("Stack:", stack)
print("Visited:", visited)
print("Parent:", parent)
print("Step:", step)
```

```python
# DFS loop
while stack:
    input("\nPress Enter to continue...")  # optional pause
    step += 1

    print(f"\n--- Step {step} ---")

    node = stack.pop()  # LIFO
    print("Popped node:", node)
    print("Stack after pop:", stack)

    if node == goal:
        print("🎯 Goal found!")
        break

    # Explore neighbors (reversed for left-to-right DFS)
    for neighbor in reversed(graph[node]):
        print(f"Checking neighbor: {neighbor}")

        if neighbor not in visited:
            visited.add(neighbor)
            parent[neighbor] = node
            stack.append(neighbor)

            print(f"  Added {neighbor} to stack")
            print("  Stack now:", stack)
            print("  Visited now:", visited)
        else:
            print(f"  {neighbor} already visited")
```

```python
def reconstruct_path(parent, start, goal):
    if goal not in parent:
        return None

    current = goal
    path = []

    while current is not None:
        path.append(current)
        current = parent[current]

    path.reverse()
    return path
```

```python
# Reconstruct and print final path
path = reconstruct_path(parent, start, goal)
print("Path from start to goal:", path)

# Optional: print parent table
print("Parent table:", parent)
```
